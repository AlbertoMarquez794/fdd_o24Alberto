```python
import requests
from bs4 import BeautifulSoup
import json
import time
import random

def get_3links(nombre):
    print(nombre)
    page = requests.get("https://www.google.com/search?q="+ nombre + " itam linkedin")
    soup = BeautifulSoup(page.content, features="lxml")
    links = soup.findAll("a")
    gLinks = []
    for i in links:
        link = i.get('href')
        if "mx.linkedin.com" in link:
            url = link.split("//")[1].split("&sa")[0]
            if len(gLinks) < 3:
                gLinks.append(url)

    return gLinks

def match_name_links(file):

    with open(file, 'r') as file:
        # Load JSON data from the file into a dictionary
        titulados_data = json.load(file)
    counter = 0
    tituladosLista = []
    # Now you can work with the dictionary
    for titulo in titulados_data:
        for espec in titulados_data[titulo]:
            listEspe = titulados_data[titulo][espec]

            for i in listEspe:
                tituladosLista.append([counter, titulo, espec] + i)
                counter += 1
                if counter <= 10:
                    print([str(counter), titulo, espec] + i)
    i = 0
    with open("AlumnosLinks.txt", "a") as file:
        while i < len(tituladosLista):
            time.sleep(5)
            try:
                x = get_3links(tituladosLista[i][3])
                if not x:
                    raise baneadoException

                z = tituladosLista[i] + x
                tituladosLista[i] = z
                print(tituladosLista[i])
                i+=1

                print("wo" + str(i))

            except baneadoException:
                print("Nos banearon")
                time.sleep(60)
            except:
                print("error")
                time.sleep(60)



    with open("AlumnosLinks.txt", "a") as file:
        # Iterate over the elements of the list
        for item in tituladosLista:
            # Write each element to the file
            file.write('\n'.join(item))

    #insertDB(tituladosLista)

class baneadoException(Exception):
    pass


match_name_links("NombresAlumnos.txt")
```
